```python
import pandas as pd
```

```python
flights = pd.read_csv("2008.csv")
flights.head()
```

```python
flights = flights[(flights["Cancelled"] == 1) & (flights["CancellationCode"] == 'B')]
flights.count()
```

```python
flights = flights[['Year', 'Month', 'Origin', 'Dest', 'Cancelled']]
flights.head()
```

```python
airports = pd.read_csv('airports.csv')
airports[airports.iata == 'LAX']
```

```python
flights = pd.merge(flights, airports[['iata', 'lat', 'long']], left_on='Origin', right_on='iata')
flights.rename(columns={'lat': 'origin_lat', 'long': 'origin_long'}, inplace = True)
flights.drop('iata', 1, inplace=True)
```

```python
flights = pd.merge(flights, airports[['iata', 'lat', 'long']], left_on='Dest', right_on='iata')
flights.rename(columns={'lat': 'dest_lat', 'long': 'dest_long'}, inplace = True)
flights.drop('iata', 1, inplace=True)
```
